```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
data = pd.read_csv('Different_stores_dataset.csv')
```

```python
print(data.head())
```

```python
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))
```

```python
#plt.subplots(rows, columns , figsize = (height, width))
fig, axarr = plt.subplots(4, 2, figsize=(16, 20))

data.groupby('region')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Region Vs Selling Price per Unit", fontsize=18)
#height spacing
plt.subplots_adjust(hspace=1.0)
#width spacing 
plt.subplots_adjust(wspace=.5)
plt.show()
```

```python

# Create subplots
fig, axarr = plt.subplots(4, 2, figsize=(12, 18))

# Plot Region Vs Selling Price per Unit
data.groupby('region')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][0], fontsize=12)
axarr[0][0].set_title("Region Vs Selling Price per Unit", fontsize=18)

# Plot Payment Method Vs Selling Price per Unit
data.groupby('payment_method')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[0][1], fontsize=12)
axarr[0][1].set_title("Payment Method Vs Selling Price per Unit", fontsize=18)

# Plot Category Vs Selling Price per Unit
data.groupby('category')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][0], fontsize=12)
axarr[1][0].set_title("Category Vs Selling Price per Unit", fontsize=18)

# Plot Gender Vs Selling Price per Unit
data.groupby('gender')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[1][1], fontsize=12)
axarr[1][1].set_title("Gender Vs Selling Price per Unit", fontsize=18)

# Plot Shopping Mall Vs Selling Price per Unit
data.groupby('shopping_mall')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[2][0], fontsize=12)
axarr[2][0].set_title("Shopping Mall Vs Selling Price per Unit", fontsize=18)

# Plot State Vs Selling Price per Unit (Top 10)
data.groupby('state')['selling_price_per_unit'].mean().sort_values(ascending=False).head(10).plot.bar(ax=axarr[2][1], fontsize=12)
axarr[2][1].set_title("State Vs Selling Price per Unit (Top 10)", fontsize=18)

# Plot Age Vs Selling Price per Unit
data.groupby('age')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][0], fontsize=12)
axarr[3][0].set_title("Age Vs Selling Price per Unit", fontsize=18)

# Plot Region Vs Selling Price per Unit
data.groupby('region')['selling_price_per_unit'].mean().sort_values(ascending=False).plot.bar(ax=axarr[3][1], fontsize=12)
axarr[3][1].set_title("Region Vs Selling Price per Unit", fontsize=18)
plt.subplots_adjust(hspace=1.0)
plt.subplots_adjust(wspace=.5)
# Display all the plots
plt.show()
```
